```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Medical Appointment.csv")
df
```

```python
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
df.head()
```

```python
# Removing unnecessary columns
df.drop(columns=["PatientId", "AppointmentID"], inplace=True)
```

```python
df.info()
```

```python
df.columns
```

```python
# Convert date columns to datetime format

df["ScheduledDay"] = pd.to_datetime(df["ScheduledDay"])
df["AppointmentDay"] = pd.to_datetime(df["AppointmentDay"])
```

```python
# It calculates the number of days between when the appointment was scheduled (ScheduledDay) and when the appointment is actually happening (AppointmentDay).
df["WaitingDays"] = (df["AppointmentDay"] - df["ScheduledDay"]).dt.days

# It extracts the name of the weekday from the AppointmentDay column.
df["AppointmentWeekDay"] = df["AppointmentDay"].dt.day_name()
```

```python
df.info()
```

```python
# Remove negative or unrealistic ages
df = df[(df['Age'] >= 0) & (df['Age'] <= 100)]
```

```python
# Encode target variable

df['No-show'] = df['No-show'].map({'Yes': 1, 'No': 0})
```

Exploratory Data Analysis

```python
sns.barplot(x ='SMS_received', y = 'No-show', data=df)
```
